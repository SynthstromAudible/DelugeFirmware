```python
import numpy as np
import matplotlib.pyplot as plt


def get_triangle_small(phase: np.uint32) -> np.int32:
    phase = np.uint32(phase)  # Ensure uint32
    if phase >= np.uint32(2147483648):  # 2^31
        phase = np.uint32(~phase + 1)  # Convert to int before negation
    return np.int32(int(phase) - 1073741824)  # Convert to int before subtraction


# Generate x values (phase values)
x = np.arange(0, 4294967296, 10000000, dtype=np.uint32)  # Cover full uint32 range

# Compute y values safely
y = np.array([get_triangle_small(p) for p in x], dtype=np.int32)
y_float = y / np.int32(2147483647)  # Normalize to float for plotting

# Create the plot
plt.plot(x, y_float, label="get_triangle_small(phase)")

# Add labels and title
plt.xlabel("Phase (uint32)")
plt.ylabel("Output (int32)")
plt.title("Plot of get_triangle_small function")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def get_triangle(phase):
    half = 1.0 / 2.0
    if phase >= half:
        phase = 1.0 - phase
    return -1 + (4 * phase)


# Generate x values (phase values)
x = np.arange(0.0, 1.0, 0.0001)  # Cover full range of [0, 1)

# Compute y values safely
y = np.array([get_triangle(p) for p in x])

# Create the plot
plt.plot(x, y, label="get_triangle(phase)")

# Add labels and title
plt.xlabel("Phase (uint32)")
plt.ylabel("Output (int32)")
plt.title("Plot of get_triangle function")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def get_triangle(phase):
    rounded = 1.0 if phase > 0.5 else 0
    return -1 + (4.0 * abs(rounded - phase))


# Generate x values (phase values)
x = np.arange(0.0, 1.0, 0.0001)  # Cover full range of [0, 1)

# Compute y values safely
y = np.array([get_triangle(p) for p in x])

# Create the plot
plt.plot(x, y, label="get_triangle(phase)")

# Add labels and title
plt.xlabel("Phase (uint32)")
plt.ylabel("Output (int32)")
plt.title("Plot of get_triangle function")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
